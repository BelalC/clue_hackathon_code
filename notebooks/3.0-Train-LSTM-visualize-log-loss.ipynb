```python
import matplotlib.pyplot as plt
%matplotlib inline

def plot_logs(history):
    """
    Plot the accuracy and loss for 
        training and test sets
    """
    evaluation_cost = history.history['val_loss']
    evaluation_accuracy = history.history['val_acc']
    training_cost = history.history['loss']
    training_accuracy = history.history['acc']
    f, (ax1, ax2) = plt.subplots(1, 2)
    f.set_figwidth(10)
    ax1.plot(evaluation_cost,label= 'validation')
    ax1.plot(training_cost, label='train')
    ax1.set_title('Log loss')
    ax1.legend()
    ax2.plot(evaluation_accuracy, label='validation')
    ax2.plot(training_accuracy, label='train')
    ax2.set_title('Accuracy')
    ax2.legend(loc='lower right')
```

```python
# 1 layer
# 100,000 train sequence, 50,000 validation sequences, step 3 days
# 15 epochs
# 16 input and output symptoms
# 90 days history
%run ../train.py
```
